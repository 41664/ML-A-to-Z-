# Deep Learning  
The input layer transfers data through synapses which have assigned weights (that is how ANNs work) and then we have the output layer. All input variables are independent to each other. We standardize the input values Output value can be numerical, binary or categorical dummy variables. 
The Neuron finds the weighted sum and applies activation function on it.
There are 4 different kinds of activation fucntion 
1. Threshhold function (Checks exactly if y=1)
2. The sigmoid fucntion or the logistic function - it is smoother than the threshhold function and tells us the probability that y=1 
3. Rectifier funtion- One of the most popular function. It is common to apply in the hidden layer. 
4. Hyperbolic tangent fucntion- The values go to below 0 so it is better.

## What actually happens during training of the Nueral Network 

1. We initialize the weightys randomly (near 0, but not 0) 
2. Input the first layer, one feature per node 
3. **Forward Propagation**- From left to right such that impact of each neuron's activation is limited by it's weight. Propagate until you get the result Y 
4. Compare predicted value to actual value and measure the generated Error 
5. Back propagate the weights from right to left and update the weights according to how much they are responsible for the error (The learning rate decides by how much we update the values)
6. Repeat 1 to 5 
7. Redo more epochs

## Problem 
A bank has a churn problem. Customers are leaving the bank. You are given a tiny sample of 10,000 customers and their details that was measured 6 months ago and you are responsible to know what is the reason of the churn and tell which customers are most likely to leave. The bank has measured after 6 months whether the customers left or are there and you have access to the data. 

```python
# Data preprocessing first 
import pandas as pd
dataset=pd.read_csv("Churn_Modelling.csv")
dataset.head()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
X=dataset.iloc[:,3:13].values
Y=dataset.iloc[:,13].values
X[0:10]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder1= LabelEncoder()
X[:,1]=labelencoder1.fit_transform(X[:,1])
X[0:10]
```

```python
# Now to avoid falling into the ummy variable trap we remove one dummy variable 
X= X[:,1:]
pd.DataFrame(X)
```

```python
X[1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train= sc.fit_transform(X_train)
X_test=sc.transform(X_test)
pd.DataFrame(X_test)
```

## Making the Artificial Neural Network 

We can define our Neural Network as a sequence or as a graph

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier= Sequential()
# Step 1 - Dense layer will take care of random initialization. 
# We chose relu for hidden and sigmoid for output 
# Initially we chose the average of input and output as the dimension of hidden layers 
#Specify the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
#The next layer knows the number of inputs
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
#Creating the output layer
# Our output is a binary value
#If there is a multiclass classification, change the activation function to softmax which is sigmoid for higher dimensions
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling Neural Network- STochastic graidentr descent on Neural NEtwork
# We need to optimize loss fucntion to get the values 
# The loss fucntion will be for the output layer that is logistic loss i.e. binary cross entropy or cateogrical cross entropy
# The metric defines the accuracy metric- this expects a list of metrics
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Give epochs and batch size
classifier.fit(X_train, Y_train, batch_size = 10, epochs = 100)
```

```python
Y_pred= classifier.predict(X_test)
Y_pred = (Y_pred > 0.5)
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
cm
```

### So this customer didn't leave the bank. Problem solved 

Now we come across the fact that repeated training of the Network yeilds different values. So we need to understand how to tune the values of the network in order to have an efficient training. In order to do so we use K-Fold cross validation which is used in the bias variance trade off understanding. 

In this we divide the model into k folds and we train on the k-1 and use the kth one as the test, k times so we can get the best bias and variance 

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
# We use this in order to use the scikit learn function with the keras model 
```

```python
# We need to define a function that builds the classifier that we need to measure
def build_classifier():
    classifier2= Sequential()
    classifier2.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier2.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier2.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier2
```

```python
# Now we need a global classifier that we can work on 
classifier_global = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
# n_jobs defines how many CPUs to use. -1 for all CPUs 
accuracies = cross_val_score(estimator = classifier_global, X=X_train, y=Y_train, cv=10, n_jobs= 2, verbose= True )
# This did not execute on my device for some odd reason.
```
